```python
import pandas as pd
```

```python
details = pd.read_csv("flight_details_atl.csv")
delays = pd.read_csv("flight_delays_atl.csv")
```

```python
print("DETAILS:")
print(details.columns)

print("\nDELAYS:")
print(delays.columns)

```

```python
common_cols = details.columns.intersection(delays.columns)
common_cols
```

```python
df_final = pd.merge(
    details,
    delays,
    on=["FL_DATE", "AIRLINE_CODE", "FL_NUMBER", "ORIGIN", "CRS_DEP_TIME"],
    how="left"
)
```

```python
details.shape, delays.shape, df_final.shape
```

### Interpretación de la unión de datos

Al realizar la unión se conservan los **153.556 registros** del dataset de detalles de vuelos. 
Se añadieron las columnas correspondientes a la información de retrasos.

Existen **2.439 vuelos** que no tienen correspondencia en el dataset de retrasos, por lo que las variables asociadas a retrasos presentan valores nulos (`NaN`), lo cual es un resultado esperado.


```python
df_final.to_csv("flights_final_atl.csv", index=False)
```
