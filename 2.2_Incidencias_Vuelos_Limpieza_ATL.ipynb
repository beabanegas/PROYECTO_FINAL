```python
ruta = r"C:\Users\beabr\Downloads\OneDrive\Formaciones\Data Analyst & AI\Proyectos\PROYECTO_FINAL\flight_delays_atl.csv"

df = pd.read_csv(ruta)
eda_preliminar(df)

df_clean = df.copy()
```

**Los cambios que vamos a realizar son:**

- FL_DATE: str --> datetime64[us]
- CRS_DEP_TIME: int64 --> HH:MM (STRING)
- CANCELLED: float64 --> 0/1 (int)
- DIVERTED: float64 --> 0/1 (int)



```python
df_clean["FL_DATE"] = pd.to_datetime(df_clean["FL_DATE"], errors="coerce")
```

```python
s = df_clean["CRS_DEP_TIME"].astype("Int64").astype(str).str.zfill(4)
df_clean["CRS_DEP_TIME"] = s.str[:2] + ":" + s.str[2:]
```

```python
df_clean["CANCELLED"] = df_clean["CANCELLED"].fillna(0).astype(int)
df_clean["DIVERTED"]  = df_clean["DIVERTED"].fillna(0).astype(int)
```
