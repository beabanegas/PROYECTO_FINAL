```python
ruta = r"C:\Users\beabr\Downloads\OneDrive\Formaciones\Data Analyst & AI\Proyectos\flight_details.csv"
df_flight_details = pd.read_csv(ruta)

df_flight_details.sample(7)
```

```python
# Función para convertir fecha (YYYY-MM-DD)
def convertir_a_fecha(valor):
    if pd.isna(valor):
        return np.nan
    try:
        return pd.to_datetime(valor, format="%Y-%m-%d").date()
    except:
        return np.nan

# Función para convertir hora HHMM → HH:MM
def convertir_a_hora(valor):
    if pd.isna(valor):
        return np.nan
    try:
        valor = int(valor)
        return f"{valor//100:02d}:{valor%100:02d}"
    except:
        return np.nan
```

```python
#1) FL_DATE (str) → convertir a fecha#
df_flight_details["FL_DATE"] = df_flight_details["FL_DATE"].apply(convertir_a_fecha)

```

```python
print("Fecha más antigua:", df_flight_details["FL_DATE"].min())
print("Fecha más reciente:", df_flight_details["FL_DATE"].max())
```
