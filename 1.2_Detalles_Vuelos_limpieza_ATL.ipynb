```python
import pandas as pd

pd.set_option('display.max_columns', None)

ruta = r"C:\Users\beabr\Downloads\OneDrive\Formaciones\Data Analyst & AI\Proyectos\PROYECTO_FINAL\flight_details_atl.csv"
df = pd.read_csv(ruta)
```

```python
# Convertir FL_DATE a fecha
df["FL_DATE"] = pd.to_datetime(df["FL_DATE"], errors="coerce")
```

```python
# CRS_DEP_TIME (int) -> "HH:MM"
s = df["CRS_DEP_TIME"].astype(int).astype(str).str.zfill(4)
df["CRS_DEP_TIME"] = pd.to_datetime(s, format="%H%M", errors="coerce")
```
